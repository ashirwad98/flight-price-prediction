```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, root_mean_squared_error
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
df=pd.read_csv(r"E:\FLIGHT PRICE PREDICTION\flight_dataset.csv")
```

```python
df
```

```python
df.drop(['Date','Month','Year','Dep_min','Arrival_min','Duration_min'],axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
le=LabelEncoder()
```

```python
lst = []
for i in df.columns:
  if df[i].dtype =='object':
    lst.append(i)
```

```python
for i in lst:
  df[i]=le.fit_transform(df[i])
  print(le.classes_)
```

```python
y=df['Price']
```

```python
x = df.drop('Price',axis = 1)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3)
```

```python
le=LinearRegression()
```

```python
le.fit(xtrain, ytrain)
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(xtrain, ytrain)
```

```python
preds = dt.predict(xtest)
```

```python
root_mean_squared_error(preds, ytest)
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(xtrain, ytrain)
```

```python
preds= rf.predict(xtest)
```

```python
root_mean_squared_error(preds, ytest)
```
